```python
from sentence_transformers import SentenceTransformer , util
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
def analyze_resume(resume,job_description):
  resume_embedding=model.encode(resume)
  job_embedding=model.encode(job_description)

  similarity=util.cos_sim(resume_embedding,job_embedding)
  score=round(float(similarity)*100,2)

  return score
```

```python
def find_missing_skills(resume, job_description):
    # Common tech skills to check
    skills = [
        "python", "javascript", "react", "node", "sql",
        "machine learning", "docker", "aws", "git",
        "html", "css", "mongodb", "typescript", "java"
    ]

    resume_lower = resume.lower()
    job_lower = job_description.lower()

    missing = []
    for skill in skills:
        if skill in job_lower and skill not in resume_lower:
            missing.append(skill)

    return missing
```

```python
resume = input("Paste your resume text: ")
job_description = input("Paste the job description: ")

score = analyze_resume(resume, job_description)
missing_skills = find_missing_skills(resume, job_description)

print("\n--- RESULTS ---")
print("Match Score:", str(score) + "%")

if score >= 75:
    print("Status: Strong Match")
elif score >= 50:
    print("Status: Moderate Match")
else:
    print("Status: Weak Match")

print("\nMissing Skills:")
if len(missing_skills) == 0:
    print("None - your resume covers all key skills!")
else:
    for skill in missing_skills:
        print("-", skill)
```
