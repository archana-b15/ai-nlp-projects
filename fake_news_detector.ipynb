```python
from transformers import pipeline
```

```python
fake_news_detector=pipeline("text-classification",model="hamzab/roberta-fake-news-classification")
```

```python
# text="According to a recent study, eating chocolate cake for breakfast every day helps you lose weight faster than dieting"
# result=fake_news_detector(text)
# print(result)
```

```python
def detect_fake_news(text):
  result=fake_news_detector(text)
  label=result[0]['label']
  score=round(result[0]['score']*100,2)
  return label,score
```

```python
user_text = input("Enter news article: ")

if len(user_text.split()) < 20:
    print("Please enter a longer text (at least 20 words)")
else:
    label, score = detect_fake_news(user_text)
    print("\n--- RESULTS ---")
    print("Verdict:", label)
    print("Confidence:", str(score) + "%")
```
