```python
from transformers import pipeline
```

```python
summariser=pipeline("summarization",model="sshleifer/distilbart-cnn-12-6")
sentiment=pipeline("sentiment-analysis",model="distilbert-base-uncased-finetuned-sst-2-english")
keyword_extract=pipeline("ner", model="yanekyuk/bert-uncased-keyword-extractor")
```

```python
def summarize(text):
  result=summariser(text,max_length=100,min_length=30)
  return result[0]['summary_text']
```

```python
def analyze_sentiment(text):
  result=sentiment(text)
  label= result[0]['label']
  score=round(result[0]['score']*100,2)
  return label,score
```

```python
def extract_keywords(text):
  result=keyword_extract(text)
  keywords=[]
  for item in result:
    keywords.append(item['word'])
  return keywords
```

```python
user_text = input("Enter your text: ")

if len(user_text.split()) < 50:
    print("Please enter a longer text (at least 50 words)")
else:
    summary = summarize(user_text)
    label, score = analyze_sentiment(user_text)
    keywords = extract_keywords(user_text)

    print("\n--- RESULTS ---")
    print("\n📝 Summary:")
    print(summary)
    print("\n😊 Sentiment:")
    print("Label:", label)
    print("Confidence:", str(score) + "%")
    print("\n🔑 Keywords:")
    print(keywords)
```
